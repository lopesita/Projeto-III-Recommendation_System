```python
df_new = df.drop(columns=['id', 'artists', 'name', 'year', 'release_date'])
```

```python
# Normalização
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_new)
```

```python
def recommend(user_input, df, df_scaled, top_n=5):
    # Procurar música ou artista no dataset
    user_rows = df[
    (df['name'].str.lower().str.contains(user_input.lower(), na=False)) |
    (df['artists'].str.lower().str.contains(user_input.lower(), na=False))]

    if user_rows.empty:
        print("Música ou artista não encontrado!")
        return

    # Calcular similaridade
    indices = user_rows.index
    similarity_scores = cosine_similarity(df_scaled[indices], df_scaled)

    # Média das similaridades (se mais de uma música/artista)
    mean_similarity = similarity_scores.mean(axis=0)

    # Obter top N similares (ignorando o input)
    top_indices = mean_similarity.argsort()[::-1]
    top_indices = [i for i in top_indices if i not in indices][:top_n]

    # Mostrar recomendações
    print(f"\nUsuário escolheu: {user_input}\nRecomendações:")
    for i in top_indices:
        print(df.iloc[i]["name"], "-", df.iloc[i]["artists"])
```

```python
user_input = input("Digite um cantor ou música de preferência: ")
recommend(user_input, df, df_scaled, top_n=5)
```
